## Analisando dados de marcas

```python
# Mostra as marcas mais citadas
autos['brand'].value_counts()
```

```python
# Seleciona os 20 primeiros
autos['brand'].value_counts().index[:20]
```

```python
# Cria um dicionário para armazenar o for
top_20_marcas = {}

# Calcula o preço médio das marcas no top 20
for b in autos['brand'].value_counts().index[:20]:
    top_marcas = autos[autos['brand'] == b]
    preco_medio = top_marcas['price'].mean()
    top_20_marcas[b] = preco_medio

# Mostra o dicionário preenchido e suas chaves
for key in top_20_marcas:
    print(key, ":", top_20_marcas[key])
```

```python
# Cria um objeto de série
marca_valor = pd.Series(top_20_marcas)
print(marca_valor)
```

```python
# Cria um dicionário para armazenar o for
km_top_20 = {}

# Calcula a média de kilometros para o top 20 de marcas
for b in autos['brand'].value_counts().index[:20]:
    top_brands = autos[autos['brand'] == b]
    media_km = top_brands['kilometer'].mean()
    km_top_20[b] = int(media_km)

# Nós mostra o resultado
for key in km_top_20:
    print(key, ":", km_top_20[key])
```

```python
# Cria um objeto de série para armazenar nosso cálculo de km por marca
marca_km = pd.Series(km_top_20)
print(marca_km)
```

```python
# Cria um dataframe com o primeiro objeto de série que criamos
final_autos = pd.DataFrame(marca_valor, columns=['mean_price'])
final_autos
```

```python
# Adiciona o segundo objeto de série como uma coluna
final_autos.insert(1, "mean_km", marca_km)
final_autos
```

```python
# Organizando a visualização do mais caro para o mais barato
final_autos.sort_values('mean_price', ascending=False)
```

```python
# Calcula a correlação entre as duas colunas
final_autos['mean_price'].corr(final_autos['mean_km'])
```
